Для начала обсудим условия решения задачи:

    * Опыт работы 3 года 2 месяца…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь
def get_expirience(row):
    if(pd.isnull(row)):
        return np.nan
    if(row == "Не указано"):
        return np.nan
    years_list = ['год', 'года', 'лет']
    months_list = ['месяц', 'месяца', 'месяцев']
    word_list = row.split()[2:6]
    res = 0
    for i,v in enumerate(word_list):
        if(v in years_list):
           res += int(word_list[i-1]) * 12
        if(v in months_list):
            res += int(word_list[i-1])
    return res
    
    
df['Опыт работы (месяц)'] = df['Опыт работы'].apply(get_expirience)
df = df.drop(columns='Опыт работы')
display(df['Опыт работы (месяц)'].median())
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
def get_city(row):
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
    word_list = row.split(' , ')
    if(word_list[0] == "Москва"):
        return "Москва"
    if(word_list[0] == 'Санкт-Петербург'):
        return "Санкт-Петербург"
    if(word_list[0] in million_cities):
        return "город-миллионник"
    return "другие"

def trim_subway_info(word_list):
    res = []
    for word in word_list:
        if(" м." not in word):
            res.append(word)
    return res

def get_relocate_willing(row):
    word_list = row.split(' , ')
    word_list = trim_subway_info(word_list)
    if("не готова" in word_list[1] or "не готов" in word_list[1]):
        return False
    return True

def get_willing_trips(row):
    word_list = row.split(' , ')
    word_list = trim_subway_info(word_list)
    try:
        if("не готова" in word_list[2] or "не готов" in word_list[2]):
            return False
        return True
    except IndexError:
        return False

df['Город'] = df['Город, переезд, командировки'].apply(get_city)
df['Готовность к переезду'] = df['Город, переезд, командировки'].apply(get_relocate_willing)
df['Готовность к командировкам'] = df['Город, переезд, командировки'].apply(get_willing_trips)
df = df.drop(columns='Город, переезд, командировки')
display(df[df['Город'] == 'Санкт-Петербург'].shape[0] / df.shape[0] * 100)
display(df[(df['Готовность к командировкам'] == True) & (df['Готовность к переезду'] == True)].shape[0] / df.shape[0] * 100)
```

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь
currencies_dict = {
    "руб.": "RUB",
    "KZT": "KZT",
    "EUR": "EUR",
    "USD": "USD",
    "грн.": "UAH",
    "бел.руб.": "BYN",
    "KGS": "KGS",
    "сум": "UZS",
    "AZN": "AZN"
}
currencies = pd.read_csv('ExchangeRates.csv')
df['Обновление резюме'] = pd.to_datetime(df['Обновление резюме'], format='%d.%m.%Y %H:%M')
df['date'] = pd.to_datetime(df['Обновление резюме'].dt.date)
currencies['date'] = pd.to_datetime(currencies['date'], format='%d/%m/%y')
df['currency'] = df['ЗП'].apply(lambda x : currencies_dict.get(x.split()[1]))
df['price'] = df['ЗП'].apply(lambda x: int(x.split()[0]))

merged = df.merge(currencies, how='left', on=['currency', 'date'])
# merged.info()
merged= merged.fillna({'close': 1, 'proportion': 1})
df['ЗП (руб)'] = merged['price'] * merged['close'] / merged['proportion']
df = df.drop(columns=['currency', 'date', 'price', 'ЗП'])
df['ЗП (руб)'].median()
df['Обновление резюме'] = df['Обновление резюме'].astype(str)
```

Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# ваш код здесь

fig, ax = plt.subplots(1, 1, figsize=(8, 4))
log_age = np.log(df['Возраст'] + 1)
histplot = sns.histplot(log_age, bins=30, ax=ax)
histplot.axvline(log_age.mean(), color='k', lw=2)
histplot.axvline(log_age.mean()+ 4 *log_age.std(), color='k', ls='--', lw=2)
histplot.axvline(log_age.mean()- 3 *log_age.std(), color='k', ls='--', lw=2)
histplot.set_title('Логарифмическое распределение возраста');

def outliers_z_score_mod(df, feature, left=3, right=3, log_scale=False):
    if log_scale:
        x = np.log(df[feature]+1)
    else:
        x = df[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = df[(x < lower_bound) | (x > upper_bound)]
    cleaned = df[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned
outliers, cleaned_data = outliers_z_score_mod(df, 'Возраст', left=3,  right=4, log_scale=True)
print(outliers.shape[0])
display(outliers)
```
